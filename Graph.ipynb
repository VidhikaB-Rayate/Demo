```python
import matplotlib.pyplot as plt
import numpy as np
# Vectors
v = [-1, 3]
u = [0, 4]
v_plus_u = [-1, 7]
v_minus_u = [-1, -1]
three_v_minus_two_u = [-3, 1]

# Create the plot
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\mathbf{v}$')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='g', label=r'$\mathbf{u}$')
plt.quiver(0, 0, v_plus_u[0], v_plus_u[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\mathbf{v + u}$')
plt.quiver(0, 0, v_minus_u[0], v_minus_u[1], angles='xy', scale_units='xy', scale=1, color='orange', label=r'$\mathbf{v - u}$')
plt.quiver(0, 0, three_v_minus_two_u[0], three_v_minus_two_u[1], angles='xy', scale_units='xy', scale=1, color='purple', label=r'$3\mathbf{v} - 2\mathbf{u}$')

# Set the axes limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Add labels
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Set aspect ratio to be equal
#plt.gca().set_aspect('equal', adjustable='box')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Title
plt.title("Graph of Vectors")

# Show the plot
plt.show()
```
